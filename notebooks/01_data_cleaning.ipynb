Recover Missing Order Month from Source File

```python
month_map = {
    "January": "01",
    "February": "02",
    "March": "03",
    "April": "04",
    "May": "05",
    "June": "06",
    "July": "07",
    "August": "08",
    "September": "09",
    "October": "10",
    "November": "11",
    "December": "12"
}

def extract_month_from_source(source_file):
    if pd.isna(source_file):
        return np.nan
    
    source_file = str(source_file)
    
    for month_name, month_num in month_map.items():
        if month_name in source_file:
            year = ''.join([char for char in source_file if char.isdigit()])
            if len(year) >= 4:
                return f"{year[:4]}-{month_num}"
    
    return np.nan
```

```python
df["order_month_from_source"] = df["source_file"].apply(extract_month_from_source)

df["order_month"] = df["order_month"].replace("NaT", np.nan)
df["order_month"] = df["order_month"].fillna(df["order_month_from_source"])
```

Data Cleaning Summary

The extracted dataset was cleaned and validated before analysis. The cleaning process included:

- Checking missing values across key columns
- Checking duplicate rows and duplicate order IDs
- Converting numeric columns into proper numeric data types
- Converting order date into datetime format
- Creating `order_month` for monthly revenue analysis
- Recovering missing order month values from `source_file`
- Standardizing text columns by removing extra spaces
- Separating completed orders and cancelled orders for different analysis purposes

Completed orders are used for revenue analysis, while cancelled orders are analyzed separately to understand cancellation patterns.
